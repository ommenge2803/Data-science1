```python
import numpy as np


# 1. Create array
arr = np.array([1, 2, 3, 4, 5])
print("1. arr:", arr)


# 2. Array of zeros
zeros = np.zeros((2, 3))
print("\n2. zeros:\n", zeros)


# 3. Array of ones
ones = np.ones((3, 2))
print("\n3. ones:\n", ones)

# 4. Identity matrix
identity = np.eye(3)
print("\n4. identity:\n", identity)

# 5. Create range of numbers
range_arr = np.arange(0, 10, 2)
print("\n5. range_arr:", range_arr)


# 6. Reshape array
reshaped = np.arange(12).reshape((3, 4))
print("\n6. reshaped:\n", reshaped)


# 7. Random numbers
rand_arr = np.random.rand(2, 3)
print("\n7. rand_arr:\n", rand_arr)


# 8. Transpose
transposed = reshaped.T
print("\n8. transposed:\n", transposed)

# 9. Flatten array
flattened = reshaped.flatten()
print("\n9. flattened:", flattened)


# 10. Stack arrays vertically
a = np.array([1, 2])
b = np.array([3, 4])
v_stack = np.vstack((a, b))
print("\n10. v_stack:\n", v_stack)


# 11. Stack arrays horizontally
h_stack = np.hstack((a, b))
print("\n11. h_stack:", h_stack)


# 12. Element-wise addition
sum_arr = a + b
print("\n12. sum_arr:", sum_arr)


# 13. Element-wise multiplication
product_arr = a * b
print("\n13. product_arr:", product_arr)


# 14. Array slicing
slice_arr = arr[1:4]
print("\n14. slice_arr:", slice_arr)


# 15. Max, Min, Mean
max_val = arr.max()
min_val = arr.min()
mean_val = arr.mean()
print("\n15. max:", max_val)   
print("    min:", min_val)     
print("    mean:", mean_val)   
```
