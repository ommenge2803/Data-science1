```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Sample data
years = ['2003', '2007', '2011', '2015', '2019']
sachin_runs = [1141, 1425, 756, 0, 0]
kohli_runs = [0, 0, 1381, 1292, 1377]
sehwag_runs = [1319, 1212, 1000, 0, 0]
x = np.arange(len(years))
width = 0.25  

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(x - width, sachin_runs, width, label='Sachin')
plt.bar(x, kohli_runs, width, label='Kohli')
plt.bar(x + width, sehwag_runs, width, label='Sehwag')

plt.xlabel('Year')
plt.ylabel('Runs')
plt.title('Runs Comparison over Selected Years')
plt.xticks(x, years)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Sample Data
players = ('sachin','kholi', 'shewag')
debut_5yrs_runs = [3256, 3999, 2950]

#Plotting
plt.figure(figsize = (8,4))
plt.bar(players, debut_5yrs_runs, color=['orange', 'red', 'green'])
plt.xlabel('Total Runs')
plt.title('Total Runs in first 5 Years of Career')
plt.tight_layout()
plt.show()
```

```python
#Pie Chart

#Sample Data
labels = ('Andriod', 'Ios', "Others")
market_share = (72, 27, 1)
explode = (0.1, 0, 0)

#Plotting
plt.figure(figsize=(8,7))
plt.pie(market_share, labels=labels, startangle=90, explode=explode)
plt.title('Mobile OS Market Share')
plt.show()
```

```python
#Plotting
plt.figure(figsize=(8,4))
plt.barh(labels, market_share, color=('Green', 'yellow', 'Blue'))
plt.xlabel('Market_share %')
plt.title('Mobile OS Market Share')
plt.tight_layout()
plt.show()
```

```python
#Line Chart
x = np.linspace(0, 10, 100)
y = 3 * x + 2
plt.figure()
plt.plot( x, y )
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Line Chart: Plot Y = 3x+ 2")
plt.grid(True)
plt.show()
```

```python
#Scatter Plot With Coustomization
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
plt.figure()
plt.scatter(x_scatter, y_scatter, color='Green', s=100)
plt.grid(True)
plt.title("Scatter Plot")
plt.show() 
```

```python
#Bar Chart With Annotation
categories = ('A', 'B', 'C', 'D', 'E')
values = (10, 20 , 30, 40, 50)
plt.figure()
bars = plt.bar(categories, values, color='skyblue')
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
             str(bar.get_height()), ha='center', va='bottom')
plt.title("Bar Chart with Annotations")
plt.show()
```

```python
#Multiple Lines In one Plot
x = np.linspace(0, 2*np.pi, 100)
plt.Figure()
plt.plot(x, np.sin(x), 'r-', label='sin')    # red solid
plt.plot(x, np.cos(x), 'g--', label='cos')   # green dashed
plt.plot(x, np.exp(x/3), 'b:', label='exp')  # blue dotted
plt.legend()
plt.title("Multiple Lines")
plt.show()
```

```python
#Histogram With Custom Bins
data = np.random.rand(1000)
plt.Figure()
plt.hist(data, bins=30, color = "Orange", edgecolor = "Black")
plt.xlabel("Values")
plt.ylabel("Frequences")
plt.title("Histogram with bins")
plt.show()
```

```python
#Subplot Grid
fig, axs = plt.subplots(2, 2, figsize=(8, 8))
fig.suptitle("2x2 Subplots Grid")

# Line plot
axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title("Line Plot")

# Bar chart
axs[0, 1].bar(categories, values)
axs[0, 1].set_title("Bar Chart")

# Scatter plot
axs[1, 0].scatter(x_scatter, y_scatter, color='purple')
axs[1, 0].set_title("Scatter Plot")

# Histogram
axs[1, 1].hist(data, bins=20, color='green')
axs[1, 1].set_title("Histogram")

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
#Stack Plot
months = np.arange(1,13)
product1 = np.random.randint(3, 15, size = 12).cumsum()
product2 = np.random.randint(3, 12, size = 12).cumsum()
product3 = np.random.randint(4, 10, size = 12).cumsum()
plt.figure()
plt.stackplot(months, product1, product2, product3,
              labels = ("Product 1, Product 2, Product s3"))
plt.legend(loc ="upper left")
plt.title("Stack Plot")
plt.show
```

```python
#Pie Chart With Example
market_share = [40, 25, 20, 15]
companies = ['Company A', 'Company B', 'Company C', 'Company D']
explode = [0.1 if i == max(market_share) else 0 for i in market_share]
plt.figure()
plt.pie(market_share, labels=companies, autopct='%1.1f%%',
        explode=explode, shadow=True)
plt.title("Pie Chart with Explode")
plt.show()
```

```python
#Axis and Legend Customization
x_rand = np.linspace(0, 10, 100)
y_rand = np.random.rand(100)
plt.figure()
plt.plot(x_rand, y_rand, label='Random Line')
plt.gca().invert_xaxis()  # Reverse X-axis
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=12)
plt.title("Axis and Legend Customization")
plt.show()
```
